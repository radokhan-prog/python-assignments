```python
#2
data = input("Enter list elements: ").split()
reversed_data = data[::-1]
print("Reversed list:", reversed_data)
```

```python
#3
nums = [int(x) for x in input("Enter numbers: ").split()]
largest = nums[0]
for n in nums:
    if n > largest:
        largest = n
print("Largest number:", largest)
```

```python
#4
data = input("Enter list elements: ").split()
if data:
    rotated = [data[-1]] + data[:-1]
    print("Rotated list:", rotated)
```

```python
#5
text = input("Enter a string: ")
word = input("Word to delete: ")
new_text = text.replace(word, "").replace("  ", " ").strip()
print("Result:", new_text)
```

```python
#6
import datetime
date_str = input("Enter date (mm/dd/yyyy): ")
date_obj = datetime.datetime.strptime(date_str, "%m/%d/%Y")
print(date_obj.strftime("%B %d, %Y"))
```

```python
#7
def capitalize_words(s):
    return ' '.join(word.capitalize() for word in s.split())

text = input("Enter a sentence: ")
print(capitalize_words(text))
```

```python
#8
rows = int(input("Enter number of rows: "))
matrix = [list(map(int, input(f"Row {i+1}: ").split())) for i in range(rows)]

for i, row in enumerate(matrix):
    print(f"Sum of row {i+1} = {sum(row)}")
```

```python
#9
r, c = map(int, input("Enter rows and columns (e.g. 2 2): ").split())
print("Enter Matrix 1:")
m1 = [list(map(int, input().split())) for _ in range(r)]
print("Enter Matrix 2:")
m2 = [list(map(int, input().split())) for _ in range(r)]

res = [[m1[i][j] + m2[i][j] for j in range(c)] for i in range(r)]
print("Resultant Matrix:")
for row in res: print(row)
```

16) Write a program to check if a year is leap year or not.

If a year is divisible by 4 then it is leap year but if the year is century year like 2000, 1900, 2100 then it must be divisible by 400.

```python
year = 2000
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(f"Is {year} a leap year? {is_leap}")
```

6) Write a program to calculate the electricity bill (accept number of unit from user) according to the following criteria :
     Unit                                                     Price  
uptp 100 units                                             no charge
Next 200 units                                              Rs 5 per unit
After 200 units                                             Rs 10 per unit
(For example if input unit is 350 than total bill amount is Rs.3500
(For example if input unit is 97 than total bill amount is Rs.0
(For example if input unit is 150 than total bill amount is Rs.750

```python
def calculate_bill():
    try:
        units = int(input("Enter the number of units: "))
        bill_amount = 0

        if units <= 100:
            # First 100 units: No charge
            bill_amount = 0

        elif units <= 300:
            # Next 200 units: Rs 5 per unit
            # We subtract the first 100 free units
            bill_amount = (units - 100) * 5

        else:
            # After 300 total units (100 free + 200 at Rs 5): Rs 10 per unit
            # Calculation: (First 100: 0) + (Next 200: 1000) + (Remainder: 10 per unit)
            remaining_units = units - 300
            bill_amount = (200 * 5) + (remaining_units * 10)

        print(f"Total bill amount is Rs.{bill_amount}")

    except ValueError:
        print("ERROR: Please enter a valid numerical value for units.")
```

13) Take input of age of 3 people by user and determine oldest and youngest among them.

```python
def compare_ages():
    try:
        # Taking inputs for three people
        age1 = int(input("Enter age of person 1: "))
        age2 = int(input("Enter age of person 2: "))
        age3 = int(input("Enter age of person 3: "))

        # Using built-in functions to find extreme values
        oldest = max(age1, age2, age3)
        youngest = min(age1, age2, age3)

        # Output the results
        print(f"\nResults:")
        print(f"The oldest person is {oldest} years old.")
        print(f"The youngest person is {youngest} years old.")

    except ValueError:
        print("ERROR: Please enter valid integers for ages.")
```
